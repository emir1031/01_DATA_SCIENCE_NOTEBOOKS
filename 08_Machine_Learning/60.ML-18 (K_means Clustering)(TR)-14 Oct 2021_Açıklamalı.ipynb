# Choosing The Optimal Number of Clusters

## Elbow metod

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 
```

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

***LK en keskin dn*** durduu noktann balangc. yani yukarda 78.85 inertia ya denk gelen 2. nokta

```python
df_diff.plot(kind='bar')
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show();
```

## Silhouette analysis

**inertia** kmelerin kendi iinde ne kadar youn biimde kmelendiini gsterirken,<br>
**silhouette** ise hem kendi iinde ne kadar sk kmelendiinin yannda kmeler arasndaki mesafenini de ne kadar uzak olduunu gsteriyor.

silhouette skor 1'e ne kadar yaknsa kmelemeye o kadar yatkn olduunu gsteriyor.
fakat bazen 1 e en yakn olan skor deil ondan biraz daha dk olan skor seilmeli nk istediim ey hem inertia skoru en dk olurken bunun yannda silhouette skorunun yksek olmas.

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, K_means_model.labels_)
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=3, random_state=42)
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```
